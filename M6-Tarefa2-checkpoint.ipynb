Considere que "próximo" neste caso significa que o valor resultante da f não pode ser maior, em módulo do que 1. Isto é $-1<f(x) - 10 < 1$, ou seja,  $9 <f(x)<11$

**Explicação:**

A malha do gráfico representa x1 vs x2.

É preciso extrair os valores em que $x1^2 + x2^2$ seja aproximadamente igual a 10, ou seja, os pares(x1,x2) com f ~ 10.

Aproximadamente significa que -1 < f(x) < 1, que é o mesmo que dizer que 9 < f(x) < 11.


Obs.: Todos os códigos são correspondentes aos códigos vistos em aula!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funções extraídas diretamente da aula

def f(x1, x2):
    return x1**2 + x2**2

def create_meshgrid(x1_lims, x2_lims):
    x1mesh, x2mesh = np.meshgrid(x1_lims, x2_lims)
    return np.c_[x1mesh.ravel(), x2mesh.ravel()]
```

```python
# Defina tamanho dos vetores para obter uma malha contínua:

x1_lims =  np.arange(-10,10,0.5)
x2_lims =  np.arange(-10,10,0.5)

print("tamanho dos vetores: x1 = {} e x2 = {}".format(len(x1_lims), len(x2_lims)))

# Crie a malha usando os vetores definidos acima:

mesh = create_meshgrid(x1_lims, x2_lims)

print("dimensoes do meshgrid: ", mesh.shape)
```

```python
# Aplique função em todos os pontos da malha e salve os valores em listas:

L = [] #lista para salvar valores

x1_w=[] #x1 para circulo branco
x2_w=[] #x2 para circulo branco

for v in mesh:
    x1 = v[0]
    x2 = v[1]
    value_f = f(x1, x2)
    if 9 < value_f < 11:        
        x1_w.append(x1)
        x2_w.append(x2)
    
    L.append(value_f)
```

```python
plt.figure(figsize=[15,10])
plt.scatter(mesh[:, 0], mesh[:, 1], c=L, cmap=plt.cm.rainbow)
plt.scatter(x1_w,x2_w, c='white')
plt.colorbar()
plt.show()
```

**O que significa o círculo branco?**
O resultado da variancia, devido nossa variavel categorica.

**Qual o objetivo desse gráfico?**

A ideia é descobrir se uma variavel é diretamente proporcional a outra.
