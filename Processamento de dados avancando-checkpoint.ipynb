### Caracteristicas não numericas:
Identificar entradas indesejadas ou registrar erros em recursos não numéricos é um pouco mais complicado. Dado que, neste momento, queremos apenas investigar a qualidade geral do conjunto de dados. Então, o que podemos fazer é dar uma olhada geral em quantos valores únicos cada uma dessas características não numéricas contêm, e com que frequência sua categoria mais frequente é representada.

```python
df_X.describe(exclude=["number", "datetime"])
```

```python
# Cria objeto de figura com 3 subtramas
fig, axes = plt.subplots(ncols=1, nrows=3, figsize=(12, 8))

# Identifique recursos não numéricos
df_non_numerical = df_X.select_dtypes(exclude=["number", "datetime"])

# Percorra os recursos e coloque cada subtrama em um objeto de eixo matplotlib
for col, ax in zip(df_non_numerical.columns, axes.ravel()):

    # Seleciona um único recurso e conta o número de valor único
    df_non_numerical[col].value_counts().plot(

       # Plota esta informação em uma figura com eixo y em escala logarítmica
        logy=True, title=col, lw=0, marker=".", ax=ax)
    
plt.tight_layout();
```

Cavando um pouco mais fundo (ou seja, olhando para as características individuais deste acidente), pudemos identificar que este acidente aconteceu em 24 de fevereiro de 2015 às 11:55 em Cardiff UK. Uma rápida pesquisa na internet revela que essa entrada corresponde a um acidente não letal, incluindo um micro-ônibus cheio de aposentados.

```python
# Coletar valores de entrada dos 10 acidentes mais frequentes:
accident_ids = df_non_numerical["Accident_Index"].value_counts().head(10).index

# Remove acidentes da lista 'accident_ids'
df_X = df_X[~df_X["Accident_Index"].isin(accident_ids)]
df_X.shape
```

### Distribuição dos Recursos
Olhar para a distribuição de valor de cada recurso é uma ótima maneira de entender melhor o conteúdo de seus dados.
Além disso, ele pode ajudar a orientar seu EDA, e fornece muitas informações úteis no que diz respeito à limpeza de dados e transformação de recursos. A maneira mais rápida de fazer isso para características numéricas é usando tramas histogramas.

```python
# Plota o histograma para cada recurso numérico em uma subtrama, separado:
df_X.hist(bins=25, figsize=(15, 25), layout=(-1, 5), edgecolor="black")
plt.tight_layout();
```

```python
# Coleta para cada recurso a entrada mais frequente
most_frequent_entry = df_X.mode()

# Verifica para cada entrada se contém a entrada mais frequente
df_freq = df_X.eq(most_frequent_entry.values, axis=1)

# Calcula a média da ocorrência 'is_most_frequent'
df_freq = df_freq.mean().sort_values(ascending=False)

# Mostre os 5 principais recursos com a maior proporção de conteúdo de valor singular
display(df_freq.head())

# Visualize the 'df_freq' table
df_freq.plot.bar(figsize=(15, 4));
```

### 3.2. Padrões de características
O próximo passo da lista é a investigação de padrões específicos de recursos. 
        O objetivo desta parte é duas vezes:
        
        Podemos identificar padrões específicos dentro de um recurso que nos ajudará a decidir se algumas entradas precisam ser descartadas ou modificadas?
        Podemos identificar relações particulares entre recursos que nos ajudarão a entender melhor nosso conjunto de dados?

```python
# Cria máscara para identificar recursos numéricos com pelo menos 25 recursos exclusivos
cols_continuous = df_X.select_dtypes(include="number").nunique() >= 25
```

### 3.2.1. Recursos contínuos
Agora que temos uma maneira de selecionar os recursos contínuos, vamos em frente e usar o seaborn's para visualizar as relações entre esses recursos.

```python
# Create a new dataframe which only contains the continuous features
df_continuous = df_X[cols_continuous[cols_continuous].index]
df_continuous.shape
```

```python
import seaborn as sns

sns.pairplot(df_continuous, height=1.5,plot_kws={"s": 2, "alpha": 0.2});
```

```python
# Parece haver uma estranha relação entre algumas características no canto superior esquerdo.
# vamos verificar
sns.pairplot(
    df_X,
    plot_kws={"s": 3, "alpha": 0.2},
    hue="Police_Force",
    palette="Spectral",
    x_vars=["Location_Easting_OSGR", "Location_Northing_OSGR", "Longitude"],
    y_vars="Latitude");
```

```python
# Crie um novo dataframe que não contenha recursos contínuos
df_discrete = df_X[cols_continuous[~cols_continuous].index]
df_discrete.shape
```

Note, para espalhar os valores na direção do eixo y precisamos escolher um recurso específico (espero que informativo). Embora o recurso 'certo' possa ajudar a identificar alguns padrões interessantes, geralmente qualquer recurso contínuo deve fazer o truque. O principal interesse neste tipo de enredo é ver quantas amostras cada valor discreto contém.

```python
# Estabeleça o número de colunas e linhas necessárias para plotar todos os recursos
n_cols = 5
n_elements = len(df_discrete.columns)
n_rows = np.ceil(n_elements / n_cols).astype("int")

# Especifique y_value para espalhar dados (idealmente um recurso contínuo)
y_value = df_X["Age_of_Driver"]

# Crie um objeto de figura com quantas linhas e colunas forem necessárias
fig, axes = plt.subplots(ncols=n_cols, nrows=n_rows, figsize=(15, n_rows * 2.5))

# Percorra os recursos e coloque cada subtrama em um objeto de eixo matplotlib
for col, ax in zip(df_discrete.columns, axes.ravel()):
    sns.stripplot(data=df_X, x=col, y=y_value, ax=ax,
                  palette="tab10", size=1, alpha=0.5)
plt.tight_layout();
```

```python
# Especificar recursos de interesse, 6 intens selecionados:
selected_features = ["Vehicle_Reference_df_res", "Towing_and_Articulation",
                     "Skidding_and_Overturning", "Bus_or_Coach_Passenger",
                     "Pedestrian_Road_Maintenance_Worker", "Age_Band_of_Driver"]

# Crie uma figura com subtramas 3 x 2
fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(16, 8))

# Faça um loop por esses recursos e plote as entradas de cada recurso em relação ao `Latitude`
for col, ax in zip(selected_features, axes.ravel()):
    sns.stripplot(data=df_X, x=col, y=df_X["Latitude"], ax=ax,
                  palette="tab10", size=2, alpha=0.5)
plt.tight_layout();
```

```python
# Crie uma figura com subtramas 3 x 2
fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(16, 8))

# Faça um loop por esses recursos e plote as entradas de cada recurso em relação ao `Latitude`
for col, ax in zip(selected_features, axes.ravel()):
    sns.violinplot(data=df_X, x=col, y=df_X["Latitude"], palette="Set2",
                   split=True, hue="Urban_or_Rural_Area", ax=ax)
plt.tight_layout();
```

```python
# Calcula a correlação de recursos
df_corr = df_X.corr(method="pearson")

# Criar rótulos para a matriz de correlação
labels = np.where(np.abs(df_corr)>0.75, "S",
                  np.where(np.abs(df_corr)>0.5, "M",
                           np.where(np.abs(df_corr)>0.25, "W", "")))

# Matriz de correlação de plotagem
plt.figure(figsize=(15, 15))
sns.heatmap(df_corr, mask=np.eye(len(df_corr)), square=True,
            center=0, annot=labels, fmt='', linewidths=.5,
            cmap="vlag", cbar_kws={"shrink": 0.8});
```

```python
#  Cria uma máscara para remover a diagonal e o triângulo superior.
lower_triangle_mask = np.tril(np.ones(df_corr.shape), k=-1).astype("bool")

#  Empilhe todas as correlações, após aplicar a máscara
df_corr_stacked = df_corr.where(lower_triangle_mask).stack().sort_values()

#  Mostrando as correlações mais baixas e mais altas na matriz de correlação
display(df_corr_stacked)
```
